```python
#Importing Librarires

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Dataset

df = pd.read_csv("upi_india_monthly_enriched.csv")

df.head()
```

```python
#Understanding the datset

df.info()
```

```python
df.describe()
```

```python
#Checking missing values

df.isnull().sum()
```

```python
df = df.dropna()
```

```python
#Converting date column to datetime format

df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.head()
```

```python
#Upi transactions volume growth

plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Volume_Mn"])

plt.title("UPI Transaction Volume Growth in India")
plt.xlabel("Year")
plt.ylabel("Transaction Volume (Millions)")

plt.show()



#Insights:
#UPI transaction volume has increased significantly over the years, showing strong -
# adoption of digital payments in India.
```

```python
#Transaction value growth

plt.figure(figsize=(10,5))

plt.plot(df["Date"], df["Value_Cr"])

plt.title("UPI Transaction Value Growth")
plt.xlabel("Year")
plt.ylabel("Transaction Value (Crores ₹)")

plt.show()



#Insights:
#As UPI usage increases, the overall transaction value also increases.
```

```python
#Yearly transaction analysis

yearly_data = df.groupby("Year")[["Volume_Mn","Value_Cr"]].sum()

yearly_data
```

```python
yearly_data.plot(kind="bar", figsize=(10,5))

plt.title("Yearly UPI Transactions Growth")

plt.show()


#Insights:
#UPI adoption has grown rapidly year by year.
```

```python
#Correlation between volume and value

corr = df[["Volume_Mn","Value_Cr"]].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Between Transaction Volume and Value")

plt.show()


#Insights:
#Higher transaction volume usually results in higher transaction value.
```

```python
#Monthly transaction pattern

monthly_data = df.groupby("Month_Name")["Volume_Mn"].mean()

monthly_data.plot(kind="bar", figsize=(10,5))

plt.title("Average Monthly UPI Transactions")

plt.show()


#Insights:
#Certain months show higher transaction volumes due to festivals and shopping seasons.
```

```python
#Covid period analysis

sns.boxplot(x=df["Is_Covid_Period"], y=df["Volume_Mn"])

plt.title("UPI Transactions During Covid")

plt.show()


#Insights:
#Digital payment usage increased during the COVID period due to reduced cash transactions.
```
